# Standardization

```python
df['room_type'] = df['room_type'].str.strip()
df
```

```python
df['price'] = df['price'].fillna(0).astype(float)
```

```python
df['price'] = df['price'].round(0).astype(float)
```

```python
df['price'] = df['price'].astype(int)

# 5. Outlier Detection
# Example using the 'price' column
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1
upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR

# Filter out outliers
df = df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)]
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def plotPerColumnDistribution(df, nGraphShown, nGraphPerRow):
    nunique = df.nunique()
    df = df[[col for col in df if nunique[col] > 1 and nunique[col] < 75]] # For displaying purposes, pick columns that have between 1 and 75 unique values
    nRow, nCol = df.shape
    columnNames = list(df)
    nGraphRow = (nCol + nGraphPerRow - 1) // nGraphPerRow
    plt.figure(num = None, figsize = (6 * nGraphPerRow, 8 * nGraphRow), dpi = 80, facecolor = 'w', edgecolor = 'k')
    for i in range(min(nCol, nGraphShown)):
        plt.subplot(nGraphRow, nGraphPerRow, i + 1)
        columnDf = df.iloc[:, i]
        if (not np.issubdtype(type(columnDf.iloc[0]), np.number)):
            valueCounts = columnDf.value_counts()
            valueCounts.plot.bar()
        else:
            columnDf.hist()
        plt.ylabel('counts')
        plt.xticks(rotation = 90)
        plt.title(f'{columnNames[i]} (column {i})')
    plt.tight_layout(pad = 1.0, w_pad = 1.0, h_pad = 1.0)
    plt.show()
# Plot the distribution of each column
plotPerColumnDistribution(df, 10, 5)
```

```python
def z_score(df, col):
    z_mean = df[col].mean()
    std = df[col].std()
    z_scores = (df[col] - z_mean) / std  # Use z_mean instead of mean
    return z_scores

# Find outliers in the 'price' column using the Z-score method
z_scores = z_score(df, 'price')
outliers_zscore = df['price'][abs(z_scores) > 3]
print(f"Outliers in 'price' using Z-score method: \n{outliers_zscore}")
```

```python
#For removal of outliers
df = df[abs(z_scores)<=3]
df
```
