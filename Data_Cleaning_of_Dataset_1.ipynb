# Handle Missing Values

```python
df['reviews_per_month'].fillna(0, inplace=True)
df['last_review'].fillna('Unknown', inplace=True)
df
```

---
# Duplicate Removal

```python
df.drop_duplicates(inplace=True)
df
```

---
# Standardization

```python
df['room_type'] = df['room_type'].str.strip()
df
```

```python
df['price'] = df['price'].fillna(0).astype(float)
```

```python
df['price'] = df['price'].round(0).astype(float)
```

```python
df['price'] = df['price'].astype(int)

# 5. Outlier Detection
# Example using the 'price' column
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1
upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR

# Filter out outliers
df = df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)]

# 6. Additional Data Cleaning
# 6.1. Check for invalid values in specific columns
# Example for 'availability_365'
df['availability_365'] = df['availability_365'].apply(lambda x: x if (x >= 0 and x <= 365) else None)
# 6.2. Data Imputation using other columns
```

Visualization

```python
# Visualize the data with a box plot
plt.figure(figsize=(14, 7))
plt.boxplot(df['price'], vert=False, patch_artist=True, showmeans=True)
plt.xlabel('Price')
plt.title('Box Plot of Price')
plt.show()

```

```python
import matplotlib.pyplot as plt

def plotPerColumnDistribution(df, num_cols=3, num_rows=2):
  fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, 6))
  axes = axes.flatten()

  for i, column in enumerate(df.columns):
    if i >= len(axes):  # Break the loop if we reach the end of the axes array
      break
    if df[column].dtype == 'object':
      # Code to handle categorical data (e.g., bar chart)
       axes[i].set_title(column)
    else:
      df[column].hist(ax=axes[i])  # Histogram for numerical data
      axes[i].set_title(column)

  plt.tight_layout()
  plt.show()


plotPerColumnDistribution(df, num_cols=3, num_rows=2)
```

```python
z_scores = (df['price'] - df['price'].mean()) / df['price'].std()
# Identify outliers
outliers = df[abs(z_scores) > 3]
print("Outliers:")
print(outliers[['name', 'price']])
```

```python
#For removal of outliers
df = df[abs(z_scores)<=3]
df
```
